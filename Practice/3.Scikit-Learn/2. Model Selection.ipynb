```python
#사이킷런의 model_selection 모듈은 
# 1. 학습 데이터와 데이터 세트를 분리
# 2. 교차 검증 분할 및 평가
# 3. Estimator(fit, predict)의 하이퍼 파라미터를 튜닝
```

```python
# K Fold 교차 검증
# 데이터를 K개로 나누고, 각각 한 번씩 총 K번 검증함
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state = 16)

# 5개의 폴드 세트로 분리하는 KFold 객체,
# 폴드 세트별 정확도를 담을 리스트 객체 생성
kfold = KFold(n_splits=5)
cv_accuracy = []
print('붓꽃 데이터 세트 크기: ',features.shape[0])
```

```python
n_iter = 0

# KFold 객체의 split()을 호출하면
# 폴드 별 학습용, 검증용 데이터의 로우 인덱스를 array로 반환함
# 각각 [0 1 2 3 4 5 ... ], [30 31 32 .....] 와 같은 형태임
# 따라서 ndarray[[0 1 2 3 4 ..]]에서는 각각 학습용, 검증용 데이터가 반환됨
for train_index, test_index in kfold.split(features):
    # kfold.split()으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    # 학습 및 예측
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter +=1

    # 반복마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test,pred),4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]

    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기 : {3}'.format(
        n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스: {1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)

print('\n## 평균 검증 정확도: ',np.mean(cv_accuracy))
```

```python
# 만약 불균형한 분포도를 가진 레이블 데이터 집합의 경우(특정 레이블 값이 특이하게 많거나 적어 한쪽으로 치우쳐진 경우)
# KFold를 사용하면 레이블 값을 제대로 반영하지 못하는 경우가 발생할 수 있음
# 따라서 원본 데이터와 유사한 대출 사기 레이블 값의 분포를 학습/테스트 세트에 유지하는게 중요함

# Stratified K폴드는 원본 데이터의 레이블 분포를 고려하여
# 이 분포와 동일하게 학습/검증 데이터 세트를 분배함으로써 이를 해결함
# K폴드와는 다르게 레이블의 분포를 고려하므로, 레이블에 대한 데이터도 인자로 받음

from sklearn.model_selection import StratifiedKFold

dt_clf = DecisionTreeClassifier(random_state = 156)

skfold = StratifiedKFold(n_splits =3)
n_iter = 0
cv_accuracy = []

# 특성 데이터 + 레이블 데이터 입력
for train_index, test_index in skfold.split(features, label):
    # kfold.split()으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    # 학습 및 예측
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter +=1

    # 반복마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test,pred),4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]

    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기 : {3}'.format(
        n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스: {1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)

print('\n## 평균 검증 정확도: ',np.mean(cv_accuracy))
```

```python
# cross_val_score()는 폴드 세트 지정, 학습 및 테스트 데이터 추출, 반복적인 학습과 예측 및 성능 반환을
# 한꺼번에 수행해주는 API임. 
# classifier -> Stratified K fold 방식, regression -> K fold 방식
# scoring 파라미터로 지정된 성능 지표 측정값을 배열 형태로 반환함

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)

features = iris_data.data
label = iris_data.target

# (estimaotr, feature data, label data, scoring = 예측 성능 평가 지표, cv = 교차검증 폴드 수)
# scoring = 'accuracy' : 성능 평가 지표는 정확도, cv = 3 : 교차 검증 세트는 3개
scores = cross_val_score(dt_clf, features, label, scoring = 'accuracy', cv=3)
print('교차 검증 별 정확도: ', np.round(scores, 4))
print('교차 검증 별 정확도: ', np.round(np.mean(scores), 4))
```

```python
# GridSearchCV는 교차검증과 하이퍼파라미터 최적화를 동시에 시행하는 API임
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd

# 데이터를 로딩하고, 학습 데이터와 테스트 데이터 분리
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target,
                                                    test_size=0.2, random_state=121)
dtree = DecisionTreeClassifier()

# DecisionTreeClassifier의 중요 하이퍼 파라미터인 
# max_depth와 min_samples_split의 값을 딕셔너리 형태로 설정
parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}

# (estimator: classifier, regressor 등, param_grid: key + list 형태의 딕셔너리가 주어짐, 
#  cv: 교차검증을 위해 분할되는 학습/테스트 세트의 개수
#  scoring: 예측 성능을 측정할 평가 방법 지정, refit: 최적의 파라미터를 찾은 뒤 재학습)
# 즉, 아래는 param_grid의 하이퍼 파라미터를 3개의 train, test set fold로 나누어 테스트 수행 설정
grid_dtree = GridSearchCV(dtree, param_grid= parameters, cv=3, scoring = 'accuracy', refit=True)

# 학습 데이터를 통해 param_grid의 하이퍼 파라미터를 순차적으로 학습/평가
grid_dtree.fit(X_train, y_train)

# GridSearchCV 결과를 추출
# parameter는 총 3x2 이므로 6개의 결과가 나타나며
# cv=3이므로 3개의 교차 검증에 대한 score(accuracy)가 표시될 수 있음
scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params','mean_test_score','rank_test_score','split0_test_score','split1_test_score','split2_test_score']]
```

```python
print('GridSearchCV 최적의 파라미터: ', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도:{0:.4f} '.format(grid_dtree.best_score_))

# refit=True이므로 GridSearchCV가 최적 성능을 나타내는 하이퍼 파라미터로
# Estimator를 학습하여 best_estimator로 저장함
# GridSearchCV의 refit으로 이미 학습된 estimator 반환
estimator = grid_dtree.best_estimator_

# GridSearchCV의 best_estimator_은 이미 최적 학습이 되어있으므로 별도 학습이 필요 없음
pred = estimator.predict(X_test)
print('테스트 데이터 세트 정확도 {0:.4f}'.format(accuracy_score(y_test,pred)))
```
