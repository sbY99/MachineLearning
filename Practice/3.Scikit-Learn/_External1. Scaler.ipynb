```python
# 분석시에 변수들의 단위 차이로 인해 숫자의 스케일이 크게 다른 경우
# 데이터 셋의 값이 들쑥날쑥 하거나 매우 큰 경우에는
# cost 값이 발산하기 때문에
# 스케일링 활용
```

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# (200개, 생성할 클러스터의 수, 클러스터 표준편차)
X, _ = make_blobs(n_samples = 200, centers = 5,
                  random_state=4, cluster_std=1.5)

print(X[0:3])
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X_train, X_test = train_test_split(X, random_state=5, test_size=0.1)

# 평균을 0, 분산을 1로 조정함
scaler = StandardScaler()

# 메소드 체이닝을 활용하여 fit, transform 연속 호출함
# 이때, 훈련용 데이터로 fit한 값을
# 테스트 데이터에도 그대로 적용해야 함 (혹은 전체 데이터에 적용 후 transform)
X_scaled = scaler.fit(X_train).transform(X_train)

X_test_scaled = scaler.transform(X_test)
plt.scatter(X_scaled[:,0],X_scaled[:,1])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 모든 값이 0~1 사이에 존재
# 정규화 방법중 원데이터 분포를 유지하면서 정규화함
scaler = MinMaxScaler()
scaler.fit(X_train)

# train 데이터로 fit한 것을 transform으로 동일하게 적용
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

plt.scatter(X_train_scaled[:,0],X_train_scaled[:,1])
plt.show()
```
