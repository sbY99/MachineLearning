```python
from sklearn.datasets import load_iris

#트리 기반 머신러닝 알고리즘을 구현한 클래스의 모임
from sklearn.tree import DecisionTreeClassifier

#학습 데이터와 검증 데이터, 예측 데이터로 데이터를 분리하거나
#최적의 하이퍼 파라미터로 평가하기 위한 모듈의 모임
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
#붓꽃 데이터 세트를 로딩
iris = load_iris()

#iris.data는 Iris 데이터 세트에서 
#특징으로 구성된 데이터를 넘파이로 가지고 있음
iris_data = iris.data
print(iris_data[:5,:])
print('-'*50)

#iris.target은 붓꽃 데이터 세트에서
#레이블(결정 값) 데이터를 넘파이로 가지고 있음
iris_label = iris.target
print(iris_label[:5])
print(iris.target_names)
print('-'*50)

iris_df = pd.DataFrame(data = iris_data, columns=iris.feature_names)
iris_df['label'] = iris_label
iris_df.head(3)
```

```python
# 학습과 테스트를 위해 iris_data를 분리함
# test_size : 테스트 데이터의 비율을 결정
# (feature 데이터 세트, label 데이터 세트, test 데이터 비율, 랜덤 난수 지정 = 숫자 맘대로)
# 반환은 튜플 형태로서, 학습용 피처 데이터, 테스트용 피처 데이터, 학습용 레이블 데이터, 테스트용 레이블 데이터 반환
x_train,x_test,y_train,y_test = train_test_split(iris_data, iris_label, test_size = 0.2, random_state =1, shuffle = True)

# 의사 결정 트리 클래스 생성
dt_clf = DecisionTreeClassifier(random_state=1)

# 학습 수행
dt_clf.fit(x_train, y_train)

# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행
pred = dt_clf.predict(x_test)
print(pred)
print('-'*50)

# 예측 성능 평가
from sklearn.metrics import accuracy_score

# (실제 레이블 데이터 세트, 예측 레이블 데이터 세트)
print('예측 정확도 : {0: .4f}'.format(accuracy_score(y_test,pred)))
```
