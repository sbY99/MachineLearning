```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
import warnings

#시각화 내용 사이즈 설정
plt.rcParams['figure.figsize'] = [7,4]

#시각화 테마 결정
sns.set(style='whitegrid')
sns.set_palette('pastel')

#warning이 많이 뜨는데, 이를 무시함
warnings.filterwarnings('ignore')

# 한글이 깨지는 경우
from matplotlib import font_manager, rc
import platform
import matplotlib

if platform.system() == 'Windows':
# 윈도우인 경우
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:    
# Mac 인 경우
    rc('font', family='AppleGothic')

matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
#Loading 'Tips' dataset from seaborn
tips = sns.load_dataset('tips')
print(tips.head())
print(tips.shape)
print(type(tips))
```

```python
#Scatter Plot
plt.scatter(tips['total_bill'], #x
            tips['tip'], #y
            color = 'pink', #테두리 색깔
            edgecolor='black', #라인 두께
            linewidth=2)

plt.show()
```

```python
#Scatter plot - seaborn
sns.scatterplot(data=tips,
                x='total_bill',
                y='tip',      
                style='time', #style: 모양 구분으로 다른 변수랑 비교
                hue = 'day',  #hue: 색깔 구분으로 다른 변수랑 비교
                size='size'  #size: 크기 구분으로 다른 변수랑 비교
                )
plt.xlabel('total_bill')
plt.ylabel('tip',fontsize=16)
plt.show()
```

```python
#Heat Map

print(tips.corr())#연속형 변수에 대한 상관관계
print('-'*50)
sns.heatmap(data = tips.corr(),
            annot=True, # 박스 안에 각각의 비율을 표시
            square=False, # 박스를 정사각형으로 할지 말지 
            vmin=-1,vmax=1, #최소 최댓값 지정
            linewidth=5,  #라인 두께 지정
            cmap='RdBu')  #색상 지정

plt.show()
```

```python
#Histogram

print(tips['total_bill'].head())
print('-'*50)

plt.hist(tips['total_bill'], #x, y축이 필요 없으므로 데이터만 넣어줌
         bins=30,            #분포를 얼마 만큼 나눌지
         density=False,      #y를 비율로 나타낼건지
         alpha=0.7,
         color='pink',
         edgecolor='black',
         linewidth=0.9)

plt.title('Histogram for total_bill')
plt.xlabel('total_bill')
plt.ylabel('sum')
plt.show()
```

```python
#seaborn

sns.histplot(data=tips,
            x='total_bill',
            bins=30,
             kde=True,          #분포를 따라 선을 그려줌
             hue='sex',
             multiple='stack',  #hue에 따라 구분된 결과를 쌓아서 표시함
             shrink=0.6)        #각 막대의 너비

plt.title('Histogram for total_bill by sex')
plt.xlabel('total_bill')
plt.ylabel('sum')
plt.show()
```

```python
#Box
plt.boxplot(tips['tip'],
            sym='rs')

plt.title('Box Plot of Tip',fontsize=20)
plt.xlabel('tip',fontsize=15)
plt.ylabel('tip size',fontsize=15)
plt.show()
```

```python
sns.boxplot(data=tips,
            x='day',
            y='tip',
            hue='smoker',
            linewidth=3,
            order=['Sun','Sat','Fri','Thur'])


plt.title('Box Plot of Tip',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Tip Size',fontsize=15)
plt.show()
```
